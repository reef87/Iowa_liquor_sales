
# Getting started

Once you've chosen your scenario, download the data from the Iowa website in csv format. Start by loading the data with pandas. You may need to parse the date columns appropriately.

```python
import pandas as pd
df = pd.read_csv('../Iowa.csv')

## Load the data into a DataFrame
# pd.read_csv()

## Transform the dates if needed, e.g.
# df["Date"] = pd.to_datetime(df["Date"], format="%m-%d-%y")
df
```

```python
# converting cost from strings into floats.
df['State Bottle Cost'] = df['State Bottle Cost'].str.replace('$','')
df['State Bottle Cost'] = df['State Bottle Cost'].astype(float)
df['State Bottle Retail'] = df['State Bottle Retail'].str.replace('$','')
df['State Bottle Retail'] = df['State Bottle Retail'].astype(float)
df['Sale (Dollars)'] = df['Sale (Dollars)'].str.replace('$','')
df['Sale (Dollars)'] = df['Sale (Dollars)'].astype(float)
```

# Build your models

Using scikit-learn or statsmodels, build the necessary models for your scenario. Evaluate model fit.

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# make dummy variables for each county
indep_dummy = pd.get_dummies(indep['County'])
```

```python
new_indep = pd.concat([df, indep_dummy], axis=1)
```

```python
new_indep.head()
```

```python
# setting up our X by dropping all of the columns we dont need (preventing overfitting)
list_to_drop = ['Date', 'Store Number', 'City', 
                'Zip Code', 'County Number', 'County',
                'Category', 'Category Name', 'Vendor Number', 
                'Item Number', 'Item Description', 
                'Bottle Volume (ml)', 'Volume Sold (Liters)',
                'Volume Sold (Gallons)', 'Sale (Dollars)']
```

```python
# save as cols not dropped
cols_not_dropped = [col for col in new_indep.columns if col not in list_to_drop]
```

```python
cols_not_dropped
```

```python
y = dep
X = new_indep[cols_not_dropped]
                   
lr = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5, random_state = 42)
lr.fit(X_train,y_train)
```

```python
new_indep[cols_not_dropped]
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
# Primary R^2 values were suspicously 
# high so I went back to investigate for over correlation
just_for_corr = X.copy()
```

```python
just_for_corr['y'] = y.values
```

```python
just_for_corr.head()
```

```python
just_for_corr.columns
```

```python
# Heatmap of correlation to check for over correlation
mean_corr = just_for_corr[['State Bottle Cost', 'State Bottle Retail', 'Bottles Sold', 'Profit', 'y']].corr()

fig, ax = plt.subplots(figsize=(9,7))

mask = np.zeros_like(mean_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

ax = sns.heatmap(mean_corr, mask=mask, ax=ax)

ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=14)
ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=14)

plt.show()
```

```python
# New model dropping Profit 
y = dep
X = new_indep[cols_not_dropped].drop('Profit', axis=1)
                   
lr = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5, random_state = 42)
lr.fit(X_train,y_train)
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```
