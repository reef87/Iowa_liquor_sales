```python
# creating a new feature for profit
df['Profit'] = (df['State Bottle Retail']-df['State Bottle Cost']) * df['Bottles Sold']
df.head()
```

```python
# this shows us the counties with the most transactions
df['County'].value_counts().head(15)
```

```python
# Creating dataframe of total transactions per county
df_top_transactions = pd.DataFrame(df['County'].value_counts())
df_top_transactions.rename(columns={'County' : 'Counts'}, inplace=True)
df_top_transactions['County'] = df_top_transactions.index
df_top_transactions.head()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=df_top_transactions['County'].head(15), 
            y=df_top_transactions['Counts'].head(15));
plt.title('Top 15 Transactions by County');
```

```python
# Sale (dolalrs) per county 
county_sales = df.groupby(df['County'])['Sale (Dollars)'].sum()
sorted_county_sales = county_sales.sort_values(ascending=False)
# Polk county has the highest sales
pd.DataFrame(sorted_county_sales)
```

```python
top_sales_county = df.groupby("County", as_index=False)["Sale (Dollars)"].sum().sort_values("Sale (Dollars)", ascending=False).head(15)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=top_sales_county["County"], 
            y=top_sales_county['Sale (Dollars)']).set(title='Sales of Top 15 Counties');
```

```python
# Counties with the highest profit margin
county_profit = df.groupby(df['County'])['Profit'].sum()
county_profit.sort_values(ascending=False).head(15)
```

```python
top_profit_county = df.groupby("County", as_index=False)["Profit"].sum().sort_values("Profit", ascending=False).head(15)
```

```python
plt.figure(figsize=(14,9))
sns.barplot(x=top_profit_county["County"], 
            y=top_profit_county['Profit']).set(title='Profit by Top 15 Counties');
```

# Building models

Using scikit-learn or statsmodels, build the necessary models for your scenario. Evaluate model fit.

```python
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# make dummy variables for each county
indep_dummy = pd.get_dummies(indep['County'])
```

```python
new_indep = pd.concat([df, indep_dummy], axis=1)
```

```python
new_indep.head()
```

```python
# setting up our X by dropping all of the columns we dont need (preventing overfitting)
list_to_drop = ['Date', 'Store Number', 'City', 
                'Zip Code', 'County Number', 'County',
                'Category', 'Category Name', 'Vendor Number', 
                'Item Number', 'Item Description', 
                'Bottle Volume (ml)', 'Volume Sold (Liters)',
                'Volume Sold (Gallons)', 'Sale (Dollars)', 'Invoice/Item Number',
                'Store Name', 'Store Location', 'Address', 'Vendor Name', 'Pack']
```

```python
# save as cols not dropped
cols_not_dropped = [col for col in new_indep.columns if col not in list_to_drop]
```

```python
cols_not_dropped
```

```python
#first run of our regression model on a 50/50 train/test set
y = dep
X = new_indep[cols_not_dropped]
                   
lr = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5, random_state = 42)
lr.fit(X_train,y_train)
```

```python
new_indep[cols_not_dropped]
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
# Primary R^2 values were suspicously 
# high so I went back to investigate for over correlation
just_for_corr = X.copy()
```

```python
just_for_corr['y'] = y.values
```

```python
just_for_corr.head()
```

```python
just_for_corr.columns
```

```python
# Heatmap of correlation to check for over correlation
mean_corr = just_for_corr[['State Bottle Cost', 'State Bottle Retail', 'Bottles Sold', 'Profit', 'y']].corr()

fig, ax = plt.subplots(figsize=(9,7))

mask = np.zeros_like(mean_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

ax = sns.heatmap(mean_corr, mask=mask, ax=ax)

ax.set_xticklabels(ax.xaxis.get_ticklabels(), fontsize=14)
ax.set_yticklabels(ax.yaxis.get_ticklabels(), fontsize=14)
plt.title('Correlation Heatmap of Variables Used in Model');
plt.show()
```

```python
# New model dropping Profit 
y = dep
X = new_indep[cols_not_dropped].drop('Profit', axis=1)
                   
lr = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5, random_state = 42)
lr.fit(X_train,y_train)
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

```python
y_preds = lr.predict(X_test)
```

```python
coeff = lr.coef_
```

```python
coeff.shape
```

```python
variable = new_indep[cols_not_dropped].drop('Profit', axis=1).columns.values
```

```python
# Creating a dataframe to see the impact a variable has on our target
coef_matrix = pd.DataFrame({'Coeff': coeff.reshape(-1), 'Variable': variable})
coef_matrix
```

```python
coef_matrix.describe()
```

```python
#looking to see which variables have the highest coefficient
#(picking out just the counties)
coef_matrix.sort_values('Coeff', ascending=False).head(30)
```
